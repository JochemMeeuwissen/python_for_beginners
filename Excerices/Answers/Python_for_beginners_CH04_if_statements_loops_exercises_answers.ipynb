```python
months = list(range(1,13))
quarters = [1,4,7,10]
years = list(range(2019,2025))

balance = 5000
recurring_deposit=500

interest = 0.1/100
bonus_interest = 0.15/100

```

```python
interest_payment = 0
quarter_counter = 0
quarter_balances = [0]
original = balance



for year in years:
    for month in months:
            
        balance = balance + recurring_deposit 
        original = original + recurring_deposit
        
        if month in quarters:
            balance = balance + interest_payment + min(quarter_balances)*bonus_interest*3/12
            interest_payment = 0 # reset
            quarter_balances = [balance]
        else:
            quarter_balances.append(balance) #collect balance for running quarter
            
        
        if not (year==2019 and month==1):
            interest_payment = interest_payment + balance*interest/12
            
        print(f"""In the {month}th month of {year}, the balance is €{balance}. The accumulated interest earnings are €{balance - original}""") if month in [1,7] else None
```
